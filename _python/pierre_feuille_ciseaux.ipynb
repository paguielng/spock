# 🎮 Pierre - Feuille - Ciseaux

Un jeu classique de Pierre-Feuille-Ciseaux avec statistiques détaillées et historique des parties.

## Comment jouer
1. Exécutez la cellule de configuration
2. Exécutez la cellule du jeu
3. Suivez les instructions interactives

```python
# 📦 Imports et Configuration
import random
from IPython.display import clear_output, display, HTML
from datetime import datetime
import json

# Constantes du jeu
ROCK = 0b001
PAPER = 0b010
SCISSORS = 0b100

# Table de victoire
winner = {
    ROCK | PAPER: PAPER,
    PAPER | SCISSORS: SCISSORS,
    SCISSORS | ROCK: ROCK,
    ROCK | ROCK: None,
    PAPER | PAPER: None,
    SCISSORS | SCISSORS: None
}

# Noms et emojis
choices_names = {
    ROCK: "Pierre 🪨",
    PAPER: "Feuille 📄",
    SCISSORS: "Ciseaux ✂️"
}

print("✅ Configuration chargée avec succès !")
```

```python
# 🎯 Classe de gestion du jeu
class PierreFeuilleCiseaux:
    def __init__(self):
        self.score = {'player': 0, 'ai': 0, 'draws': 0}
        self.history = []
        self.total_games = 0
        
    def reset_stats(self):
        """Réinitialiser les statistiques"""
        self.score = {'player': 0, 'ai': 0, 'draws': 0}
        self.history = []
        self.total_games = 0
        
    def play_round(self, player_choice, ai_choice):
        """Jouer un tour et retourner le résultat"""
        result = None
        if player_choice == winner[player_choice | ai_choice]:
            result = 'player'
            self.score['player'] += 1
        elif ai_choice == winner[player_choice | ai_choice]:
            result = 'ai'
            self.score['ai'] += 1
        else:
            result = 'draw'
            self.score['draws'] += 1
            
        # Enregistrer dans l'historique
        self.history.append({
            'round': len(self.history) + 1,
            'player': player_choice,
            'ai': ai_choice,
            'result': result,
            'timestamp': datetime.now().strftime("%H:%M:%S")
        })
        self.total_games += 1
        
        return result
    
    def display_round_result(self, player_choice, ai_choice, result):
        """Afficher le résultat d'un tour de manière stylée"""
        html = f"""
        <div style="padding: 20px; border-radius: 10px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; margin: 10px 0;">
            <h3 style="margin: 0 0 15px 0;">🎲 Tour {len(self.history)}</h3>
            <div style="display: flex; justify-content: space-around; align-items: center; font-size: 24px;">
                <div style="text-align: center;">
                    <div style="font-size: 48px;">{choices_names[player_choice].split()[1]}</div>
                    <div>Vous</div>
                </div>
                <div style="font-size: 36px;">VS</div>
                <div style="text-align: center;">
                    <div style="font-size: 48px;">{choices_names[ai_choice].split()[1]}</div>
                    <div>Ordinateur</div>
                </div>
            </div>
        """
        
        if result == 'player':
            html += '<h2 style="text-align: center; margin-top: 20px; color: #4ade80;">🎉 Vous avez gagné !</h2>'
        elif result == 'ai':
            html += '<h2 style="text-align: center; margin-top: 20px; color: #f87171;">😔 Vous avez perdu !</h2>'
        else:
            html += '<h2 style="text-align: center; margin-top: 20px; color: #fbbf24;">🤝 Égalité !</h2>'
            
        html += '</div>'
        display(HTML(html))
    
    def display_score(self):
        """Afficher le score actuel"""
        total = self.score['player'] + self.score['ai'] + self.score['draws']
        win_rate = (self.score['player'] / total * 100) if total > 0 else 0
        
        html = f"""
        <div style="padding: 20px; border-radius: 10px; background: #1f2937; color: white; margin: 10px 0;">
            <h3 style="text-align: center; margin: 0 0 15px 0;">📊 Score</h3>
            <div style="display: flex; justify-content: space-around; text-align: center;">
                <div>
                    <div style="font-size: 32px; font-weight: bold; color: #4ade80;">{self.score['player']}</div>
                    <div>Vous</div>
                </div>
                <div>
                    <div style="font-size: 32px; font-weight: bold; color: #fbbf24;">{self.score['draws']}</div>
                    <div>Égalités</div>
                </div>
                <div>
                    <div style="font-size: 32px; font-weight: bold; color: #f87171;">{self.score['ai']}</div>
                    <div>Ordinateur</div>
                </div>
            </div>
            <div style="text-align: center; margin-top: 15px; padding-top: 15px; border-top: 1px solid #374151;">
                <div>Taux de victoire : <strong>{win_rate:.1f}%</strong></div>
            </div>
        </div>
        """
        display(HTML(html))
    
    def display_history(self, last_n=10):
        """Afficher l'historique des dernières parties"""
        if not self.history:
            print("Aucune partie jouée pour le moment.")
            return
            
        recent_history = self.history[-last_n:]
        
        html = '<div style="padding: 20px; border-radius: 10px; background: #f3f4f6; margin: 10px 0;">'
        html += f'<h3 style="margin: 0 0 15px 0;">📜 Historique (derniers {len(recent_history)} tours)</h3>'
        html += '<table style="width: 100%; border-collapse: collapse;">'
        html += '<tr style="background: #e5e7eb; font-weight: bold;"><th style="padding: 8px;">Tour</th><th>Heure</th><th>Vous</th><th>Ordinateur</th><th>Résultat</th></tr>'
        
        for game in recent_history:
            result_text = "🏆 Victoire" if game['result'] == 'player' else "❌ Défaite" if game['result'] == 'ai' else "🤝 Égalité"
            result_color = "#4ade80" if game['result'] == 'player' else "#f87171" if game['result'] == 'ai' else "#fbbf24"
            
            html += f"""
            <tr style="border-bottom: 1px solid #d1d5db;">
                <td style="padding: 8px; text-align: center;">{game['round']}</td>
                <td style="padding: 8px; text-align: center;">{game['timestamp']}</td>
                <td style="padding: 8px; text-align: center;">{choices_names[game['player']]}</td>
                <td style="padding: 8px; text-align: center;">{choices_names[game['ai']]}</td>
                <td style="padding: 8px; text-align: center; color: {result_color}; font-weight: bold;">{result_text}</td>
            </tr>
            """
        
        html += '</table></div>'
        display(HTML(html))
    
    def get_statistics(self):
        """Calculer et afficher des statistiques détaillées"""
        if not self.history:
            print("Aucune statistique disponible.")
            return
        
        # Compter les choix
        player_choices = {ROCK: 0, PAPER: 0, SCISSORS: 0}
        ai_choices = {ROCK: 0, PAPER: 0, SCISSORS: 0}
        
        for game in self.history:
            player_choices[game['player']] += 1
            ai_choices[game['ai']] += 1
        
        total = len(self.history)
        win_rate = (self.score['player'] / total * 100)
        
        html = f"""
        <div style="padding: 20px; border-radius: 10px; background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); color: white; margin: 10px 0;">
            <h3 style="margin: 0 0 15px 0;">📈 Statistiques Détaillées</h3>
            <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 20px;">
                <div>
                    <h4>Vos Choix</h4>
                    <div>🪨 Pierre : {player_choices[ROCK]} ({player_choices[ROCK]/total*100:.1f}%)</div>
                    <div>📄 Feuille : {player_choices[PAPER]} ({player_choices[PAPER]/total*100:.1f}%)</div>
                    <div>✂️ Ciseaux : {player_choices[SCISSORS]} ({player_choices[SCISSORS]/total*100:.1f}%)</div>
                </div>
                <div>
                    <h4>Choix de l'Ordinateur</h4>
                    <div>🪨 Pierre : {ai_choices[ROCK]} ({ai_choices[ROCK]/total*100:.1f}%)</div>
                    <div>📄 Feuille : {ai_choices[PAPER]} ({ai_choices[PAPER]/total*100:.1f}%)</div>
                    <div>✂️ Ciseaux : {ai_choices[SCISSORS]} ({ai_choices[SCISSORS]/total*100:.1f}%)</div>
                </div>
            </div>
            <div style="margin-top: 20px; padding-top: 20px; border-top: 1px solid rgba(255,255,255,0.3);">
                <div><strong>Total de parties :</strong> {total}</div>
                <div><strong>Victoires :</strong> {self.score['player']} ({win_rate:.1f}%)</div>
                <div><strong>Défaites :</strong> {self.score['ai']} ({self.score['ai']/total*100:.1f}%)</div>
                <div><strong>Égalités :</strong> {self.score['draws']} ({self.score['draws']/total*100:.1f}%)</div>
            </div>
        </div>
        """
        display(HTML(html))

# Créer une instance du jeu
game = PierreFeuilleCiseaux()
print("✅ Classe du jeu initialisée !")
```

## 🎮 Jouer une Partie

Exécutez la cellule ci-dessous pour démarrer une nouvelle partie.

```python
# 🎲 JEU PRINCIPAL
def play_game():
    clear_output(wait=True)
    
    # Afficher le titre
    display(HTML("""
    <div style="text-align: center; padding: 30px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); 
                color: white; border-radius: 15px; margin-bottom: 20px;">
        <h1 style="margin: 0; font-size: 48px;">🎮 Pierre Feuille Ciseaux</h1>
        <p style="margin: 10px 0 0 0; font-size: 18px;">Que le meilleur gagne !</p>
    </div>
    """))
    
    # Demander le nombre de tours
    while True:
        try:
            turns = int(input("🎯 En combien de tours voulez-vous jouer ? (ex: 5) : "))
            if turns > 0:
                break
            else:
                print("⚠️ Veuillez entrer un nombre positif.")
        except ValueError:
            print("⚠️ Veuillez entrer un nombre valide.")
    
    # Boucle de jeu
    for turn in range(turns):
        clear_output(wait=True)
        
        # Afficher le score actuel
        if turn > 0:
            game.display_score()
        
        # Menu de choix
        display(HTML(f"""
        <div style="padding: 20px; border-radius: 10px; background: #f3f4f6; margin: 20px 0;">
            <h3 style="text-align: center; margin: 0 0 15px 0;">🎯 Tour {turn + 1} / {turns}</h3>
            <div style="text-align: center; font-size: 18px;">
                <p><strong>Que voulez-vous jouer ?</strong></p>
                <p>1️⃣ - Pierre 🪨</p>
                <p>2️⃣ - Feuille 📄</p>
                <p>3️⃣ - Ciseaux ✂️</p>
            </div>
        </div>
        """))
        
        # Choix du joueur
        while True:
            try:
                choice = int(input("Votre choix (1, 2 ou 3) : "))
                if choice in (1, 2, 3):
                    break
                else:
                    print("⚠️ Veuillez choisir 1, 2 ou 3.")
            except ValueError:
                print("⚠️ Veuillez entrer un nombre valide.")
        
        player_choice = [ROCK, PAPER, SCISSORS][choice - 1]
        ai_choice = random.choice([ROCK, PAPER, SCISSORS])
        
        # Jouer le tour
        result = game.play_round(player_choice, ai_choice)
        
        # Afficher le résultat
        clear_output(wait=True)
        game.display_round_result(player_choice, ai_choice, result)
        
        # Pause avant le prochain tour
        if turn < turns - 1:
            input("\nAppuyez sur Entrée pour continuer...")
    
    # Afficher le résumé final
    clear_output(wait=True)
    
    display(HTML("""
    <div style="text-align: center; padding: 30px; background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); 
                color: white; border-radius: 15px; margin-bottom: 20px;">
        <h1 style="margin: 0; font-size: 48px;">🏁 Partie Terminée !</h1>
    </div>
    """))
    
    game.display_score()
    
    # Afficher le résultat final
    if game.score['player'] > game.score['ai']:
        display(HTML('<h2 style="text-align: center; color: #4ade80;">🎉🎉🎉 Félicitations ! Vous avez gagné ! 🎉🎉🎉</h2>'))
    elif game.score['ai'] > game.score['player']:
        display(HTML('<h2 style="text-align: center; color: #f87171;">😔 Dommage ! L\'ordinateur a gagné ! Réessayez !</h2>'))
    else:
        display(HTML('<h2 style="text-align: center; color: #fbbf24;">🤝 Match nul ! Quelle partie serrée !</h2>'))
    
    print("\n💡 Vous pouvez consulter vos statistiques dans les cellules suivantes !")

# Lancer le jeu
play_game()
```

## 💾 Sauvegarde et Chargement

Sauvegardez vos statistiques pour les retrouver plus tard !

```python
# 💾 Sauvegarder les statistiques
def save_game(filename='game_stats.json'):
    data = {
        'score': game.score,
        'history': game.history,
        'total_games': game.total_games,
        'saved_at': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f, indent=2, ensure_ascii=False)
    print(f"✅ Statistiques sauvegardées dans {filename}")

def load_game(filename='game_stats.json'):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            data = json.load(f)
        game.score = data['score']
        game.history = data['history']
        game.total_games = data['total_games']
        print(f"✅ Statistiques chargées depuis {filename}")
        print(f"   Sauvegardées le : {data['saved_at']}")
    except FileNotFoundError:
        print(f"⚠️ Fichier {filename} non trouvé.")

# Sauvegarder
save_game()
```

```python
# 📂 Charger une sauvegarde
load_game()
```
